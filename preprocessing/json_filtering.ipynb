```python
#read arlington zipcodes
filename = '../data/arlington_zips.txt'

with open(filename, 'r') as a:
    z = a.readlines()[0:-1]

zips = [line.strip() for line in z]
#contents['type']
```

```python
arl_zips = {'type': contents['type']}
arl_zips["features"] = []

for feature in contents['features']:
    if feature['properties']['ZCTA5CE10'] in zips:
        print(feature['properties']['ZCTA5CE10'])
        arl_zips['features'].append(feature)
```

```python
with open("../data/arl_zips.json", "w") as outfile:
    json.dump(arl_zips, outfile)
```
